## 1. Load and Inspect

Start by reading the dataset and checking the basic structure.

```python
path = '/kaggle/input/datasets/mubashirsidiki/2026-github-most-starred-projects/github_most_starred.csv'
df = pd.read_csv(path)

df['lang_clean'] = df['lang'].fillna('Unknown')

print('shape:', df.shape)
display(df.head())

summary = pd.DataFrame({
    'missing': df.isna().sum(),
    'dtype': df.dtypes.astype(str),
})
display(summary)
```

## 2. Dataset Snapshot

A few things stand out immediately from the real values in this file:

- There are exactly 100 repositories in the leaderboard.
- `build-your-own-x` leads with 490,392 stars, followed by `awesome` and `freeCodeCamp`.
- The most common languages are Python, TypeScript, and Unknown entries where the language is missing.
- Stars are heavily concentrated in the upper range, with a median around 138,836 and a maximum near 490k.

## 3. Top Repositories and Languages

Now we look at the most starred projects, the language mix, and how stars relate to forks.

```python
fig, axes = plt.subplots(1, 3, figsize=(22, 6), gridspec_kw={'width_ratios': [1.2, 1.0, 1.1]})

repos_plot = df.sort_values('stars', ascending=False).head(10).sort_values('stars')
sns.barplot(data=repos_plot, y='repo-name', x='stars', hue='repo-name', ax=axes[0], palette='crest', dodge=False, legend=False)
axes[0].set_title('Top 10 Repositories by Stars')
axes[0].set_xlabel('Stars')
axes[0].set_ylabel('')
axes[0].grid(axis='x', alpha=0.25)
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.0f', padding=4, fontsize=9)

lang_plot = df['lang_clean'].value_counts().head(10).sort_values()
sns.barplot(x=lang_plot.values, y=lang_plot.index, hue=lang_plot.index, ax=axes[1], palette='magma', dodge=False, legend=False)
axes[1].set_title('Top Languages')
axes[1].set_xlabel('Repositories')
axes[1].set_ylabel('')
axes[1].grid(axis='x', alpha=0.25)

sns.scatterplot(
    data=df,
    x='forks',
    y='stars',
    hue='lang_clean',
    palette='tab10',
    s=90,
    alpha=0.85,
    ax=axes[2],
    legend=False,
)
axes[2].set_title('Stars vs Forks')
axes[2].set_xlabel('Forks')
axes[2].set_ylabel('Stars')
axes[2].set_xscale('log')
axes[2].set_yscale('log')
axes[2].grid(True, which='both', alpha=0.22)

for ax in axes:
    ax.tick_params(axis='y', labelsize=11)

plt.tight_layout()
plt.show()
```

## 4. Key Takeaways

- `build-your-own-x`, `awesome`, and `freeCodeCamp` anchor the top of the leaderboard.
- Python and TypeScript make up a large share of the list, but the dataset also includes Markdown, JavaScript, Rust, Go, and more.
- The stars-vs-forks plot shows a strong high-popularity cluster rather than a long tail of tiny projects.
