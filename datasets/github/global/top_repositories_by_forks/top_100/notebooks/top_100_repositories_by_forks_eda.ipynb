## 1. Load and Inspect

Load the dataset, preview records, and check missingness.

```python
from pathlib import Path
csv_name = 'top_100_repositories_by_forks.csv'
candidate_paths = list(Path('/kaggle/input').rglob(csv_name))
path = str(candidate_paths[0]) if candidate_paths else csv_name
df = pd.read_csv(path)
print('Loaded from:', path)
print('shape:', df.shape)
display(df.head())
profile = pd.DataFrame({'missing': df.isna().sum(), 'dtype': df.dtypes.astype(str)})
display(profile)
```

## 2. Dataset Snapshot

These notes are based on the current file output:

- Rows/columns: **100 x 9**
- Highest `forks` repo: **datasharing** (**242,765**)
- Highest `stars` repo: **build-your-own-x** (**490,392**)
- `forks` median: **38,884** | max: **242,765**
- `stars` vs `forks` Pearson correlation: **-0.021** (near-zero linear relationship)
- Missing values: **15**

## 3. Visual Summary

Split charts for readability.

```python
metric_col = 'forks'
top_plot = df[['project_name', metric_col]].sort_values(metric_col, ascending=False).head(12).sort_values(metric_col).copy()
top_plot['project_plot'] = top_plot['project_name'].astype(str).str.slice(0, 45)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_plot, y='project_plot', x=metric_col, hue='project_plot', dodge=False, legend=False, palette='crest')
plt.title(f'Top 12 Repositories by {metric_col}')
plt.xlabel(metric_col)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
lang_plot = df['language'].fillna('Unknown').value_counts().head(10).sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=lang_plot.values, y=lang_plot.index, hue=lang_plot.index, dodge=False, legend=False, palette='magma')
plt.title('Top Languages in the Ranking')
plt.xlabel('Repositories')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='forks', y='stars', s=70, alpha=0.75, color='#264653')
plt.xscale('log')
plt.yscale('log')
plt.title('Stars vs Forks (log-log)')
plt.xlabel('Forks (log)')
plt.ylabel('Stars (log)')
plt.tight_layout()
plt.show()
```

## 4. Key Takeaways

- The leaderboard is driven by very large `forks` values, with a wide spread across repositories.
- `stars` and `forks` show almost no linear relationship here (Pearson correlation about **-0.021**).
- Most common language here is `Python`, but this is still a mixed-language leaderboard.
