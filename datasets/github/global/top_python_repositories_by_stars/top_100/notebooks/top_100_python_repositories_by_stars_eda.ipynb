## 1. Load and Inspect

Load the dataset, preview records, and check missingness.

```python
from pathlib import Path
csv_name = 'top_100_python_repositories_by_stars.csv'
candidate_paths = list(Path('/kaggle/input').rglob(csv_name))
path = str(candidate_paths[0]) if candidate_paths else csv_name
df = pd.read_csv(path)
print('Loaded from:', path)
print('shape:', df.shape)
display(df.head())
profile = pd.DataFrame({'missing': df.isna().sum(), 'dtype': df.dtypes.astype(str)})
display(profile)
```

## 2. Dataset Snapshot

These notes are based on the current file output:

## 3. Visual Summary

Split charts for readability.

```python
metric_col = 'stars'
top_plot = df[['project_name', metric_col]].sort_values(metric_col, ascending=False).head(12).sort_values(metric_col).copy()
top_plot['project_plot'] = top_plot['project_name'].astype(str).str.slice(0, 45)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_plot, y='project_plot', x=metric_col, hue='project_plot', dodge=False, legend=False, palette='crest')
plt.title(f'Top 12 Repositories by {metric_col}')
plt.xlabel(metric_col)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
issue_plot = df[['project_name', 'open_issues']].sort_values('open_issues', ascending=False).head(12).sort_values('open_issues').copy()
issue_plot['project_plot'] = issue_plot['project_name'].astype(str).str.slice(0, 45)
plt.figure(figsize=(10, 7))
sns.barplot(data=issue_plot, y='project_plot', x='open_issues', hue='project_plot', dodge=False, legend=False, palette='magma')
plt.title('Top Repositories by Open Issues')
plt.xlabel('Open Issues')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='forks', y='stars', s=70, alpha=0.75, color='#264653')
plt.xscale('log')
plt.yscale('log')
plt.title('Stars vs Forks (log-log)')
plt.xlabel('Forks (log)')
plt.ylabel('Stars (log)')
plt.tight_layout()
plt.show()
```

## 5. Key Takeaways

- The leaderboard is driven by very large `stars` values, with a wide spread across repositories.
- `stars` and `forks` move together positively (correlation about 0.716), but not perfectly.
- Language is fully concentrated in Python in this dataset, so ranking differences come from project popularity/age/ecosystem position.
