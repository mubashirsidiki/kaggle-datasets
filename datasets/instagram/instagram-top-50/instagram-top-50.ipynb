# Top 50 Instagram Accounts by Followers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.style.use('dark_background')

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import glob
csv_path = glob.glob('/kaggle/input/**/*.csv', recursive=True)[0]
df = pd.read_csv(csv_path)
df.head()
```

```python
colors = plt.cm.RdPu(np.linspace(0.2, 0.9, 50))

fig, ax = plt.subplots(figsize=(12, 18))
bars = ax.barh(range(len(df)), range(len(df), 0, -1), color=colors[::-1], edgecolor='none', height=0.7)

for i, row in df.iterrows():
    ax.text(0.3, i, f"{row['account']}  @{row['username']}", va='center', fontsize=9,
            color='white', fontweight='bold' if i < 5 else 'normal')

ax.set_yticks([])
ax.invert_yaxis()
ax.set_title('Top 50 Instagram Accounts by Followers', fontsize=16, fontweight='bold', pad=15, color='white')
ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)
ax.xaxis.set_visible(False)

ax.text(26, 1, '#1', fontsize=11, va='center', color='#ff6eb4', fontweight='bold')
ax.text(26, 6, '#5', fontsize=10, va='center', color='#ff85c8')
ax.text(26, 16, '#15', fontsize=9, va='center', color='#aaa')
ax.text(26, 36, '#35', fontsize=9, va='center', color='#aaa')

fig.patch.set_facecolor('#0d1117')
ax.set_facecolor('#0d1117')
plt.tight_layout()
plt.show()
```
